```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
from utils import generate_table
```

```python
base_path = Path("../../experiments/cmnist/2digits/2colors/1missing/ultimate/")

subsampled_fallen_sun = {
    ("\\texttt{ZSF+bal.\\ (ranking)}", ""): "cmnist_ranking.csv",
#     ("\\texttt{ZSF+bal.\\ (ranking) - no predictor}", ""): "cmnist_ranking_no_predictors.csv",
    ("\\texttt{ZSF+bal.\\ (k-means)}", ""): "cmnist_kmeans.csv",
    ("\\texttt{ZSF+bal.\\ (ground truth)}", ""): "cmnist_perfect_cluster.csv",
    ("\\texttt{ZSF}", ""): "cmnist_no_cluster.csv",
    ("\\texttt{FWD \\cite{HasSriNamLia18}}", "eta"): "cmnist_dro_baseline_color_40epochs.csv",
    ("\\texttt{CNN}", ""): "cmnist_cnn_baseline_color_40epochs.csv",
}
```

```python
def gather(file_dict):
    df_all = pd.DataFrame()
    for (approach, param), filename in file_dict.items():
        df = pd.read_csv(base_path / filename)
        df.insert(0, "approach", approach)
        if param:
            df["approach"] = df["approach"] + " (" + df[param].astype(str) + ")"
        df_all = pd.concat([df_all, df], axis="index", ignore_index=True, sort=False)
    return df_all
df = gather(subsampled_fallen_sun)
```
