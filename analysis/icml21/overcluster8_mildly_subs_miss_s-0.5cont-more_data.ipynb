```python
wandb = RunsDownloader(project="fdm-hydra")
data_ = wandb.groups(
    "ranking-fdm-6.overcluster6.mildly_subs_miss_s",
    "kmeans-fdm-6.overcluster6.mildly_subs_miss_s",
    "ranking-fdm-8.overcluster8.mildly_subs_miss_s",
    "kmeans-fdm-8.overcluster8.mildly_subs_miss_s",
)
data_alt = simple_concat(data_alt, data_)
```

```python
plot_kwargs_alt = dict(
    file_format="pdf",
    fig_dim=(5, 2.5),
    file_prefix="cmnist_2v4_miss_s_alt_overcluster",
)
```
